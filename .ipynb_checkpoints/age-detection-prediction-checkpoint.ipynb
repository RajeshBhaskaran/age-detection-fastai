## Prediction

This notebook is fairly simple - we will load a previously trained neural network and predict the age of a person utilizing an image. The neural net was trained using FASTAI. Please refer to the notebook named age-detection-training for more information about how this model was trained. 

This model was based on the dataset by the Open University of Israel (link in the reference section) and provides images classified in eight categories of age - 
* 0-2
* 4-6
* 8-12
* 15-20
* 25-32
* 38-43
* 48-53
* 60-100

## Loading the Model

FASTAI models are exported in a single .pkl file. In this repository the file "export.pkl" provided in the folder called "models".

In the code below. Just ensure that the variable "path" reflects the path to the folder containing the "export.pkl" file. 

```python
path = "models" 
learn = load_learner(path)
```

## Using the Model

Below we will use the model for inference using some sample imagery I downloaded from Pexels.com. 
The process is quite simple - you load an image file and then use the lean.predict function to get the category. 

```python
img = open_image("images/toddler.jpeg")
img.show()
```

```python
learn.predict(img)
```

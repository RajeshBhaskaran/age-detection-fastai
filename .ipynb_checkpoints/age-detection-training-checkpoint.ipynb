## Loading the Dataset

Creating all neccesary folders in Colab to copy and unzip the dataset. 

```python
newpath = "/.root/.fastai/data/age-detection"
!mkdir -p {newpath}
path = Path(newpath)
img_path = path/"faces"
```

```python
# Make sure to update "/content/Docs/My Drive/faces.tar.gz" to be the appropriate path
!cp -v "/content/Docs/My Drive/faces.tar.gz" {path}
!7za -bd -y -so x "/content/Docs/My Drive/faces.tar.gz" | tar xf - -C {path.as_posix()}
```

I created a CSV file that maps the imagery to the age categories based on the txt files provided in the dataset. I merely stripped a lot of information I wasn't going to use (like gender for example). Also fixed a few typos I found. You do can download this CSV out of the "/models" folder of the GIT repo. 

```python
# Make sure to update "/content/Docs/My Drive/label-v4.csv" to be the appropriate path
!cp -v "/content/Docs/My Drive/label-v4.csv" {path}
labels = pd.read_csv(path/"label-v4.csv")
labels.head()
```

```python
len(labels)
```

Below the FASTAI specific code - 

```python
np.random.seed(2)
bs = 64
size = 224
```

```python
tfms = get_transforms(do_flip=True, flip_vert=False, max_rotate=0.1, max_zoom=0.1, max_lighting=0.4, 
                      max_warp=0.4,p_affine=0.3, p_lighting=0.3)
```

```python
data = ImageDataBunch.from_df(img_path,labels,ds_tfms=tfms, size=size, bs=bs).normalize(imagenet_stats)
```

```python
data.export()
```

```python
data.show_batch(rows=3, figsize=(7,6))
```

```python
data.classes
```

## Creating & Training the Neural Net

```python
learn = cnn_learner(data, models.resnet50, metrics=(error_rate,accuracy))
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
lr = 0.003 
```

```python
learn.fit_one_cycle(25, lr, pct_start=0.3)
savedModel = learn.save("age-dectection-v8", return_path=True)
!cp -v {savedModel} "/content/Docs/My Drive"
learn.export()
```

```python
learn.unfreeze()
```

```python
learn.fit_one_cycle(5, slice(1e-6,1e-4))
```

```python
savedModel = learn.save("age-dectection-v8b", return_path=True)
!cp -v {savedModel} "/content/Docs/My Drive"
learn.export()
```

```python
interp = ClassificationInterpretation.from_learner(learn)
```

```python
interp.plot_confusion_matrix(figsize=(12,12), dpi=60)
```

## Results 

The model has an accuracy of 84.4% which is really when compared with the original scores foound in Tal Hassner's paper from 2015 - https://talhassner.github.io/home/projects/cnn_agegender/CVPR2015_CNN_AgeGenderEstimation.pdf

The accuracy they reported was of 50.7%, so this model is a very important improvement. In addition the paper mentioned a "1-off" category that yielded an accuracy of 84.7%. This meant that the model failed to predict the correct category, but predicted an adjacent category (so it was not so far off). Using the confusion matrix above I calculated the "1-off" accuracy of this model to be 96.6%.
